Заметим, что хоть у нас и 10126 строк, уникальных организаций у нас 8124. Это связано с тем, что если организация сетевая, то каждая сеть заводит свой аккаунт в соцсетях. таким образом, мы можем наблюдать, что, например, у сети "Планета здоровья" 55 точек, у которых есть аккаунт в ВК, у "Столички" - 47 и так далее. Либо жто просто множество созданных аккаунтов у какой-либо аптеки, магазина и т. д.

```python


def extract_city_and_metro(address):
    parts = address.split(',')
    city = parts[0].strip() if len(parts) > 0 else None
    metro = parts[-1].strip() if len(parts) > 1 else None
    return pd.Series([city, metro])

new_columns = merged_df['Адрес места'].apply(extract_city_and_metro)
merged_df[['Город', 'Метро']] = new_columns

merged_df.head()
```

```python
import re

def extract_city_and_floor(address):
    parts = address.split(',')
    city = parts[-1].strip() if len(parts) > 1 else None
    floor_match = re.search(r'([0-9]+)', address)
    floor = floor_match.group(0) if floor_match else None
    return pd.Series([city, floor])

new_columns = cleaned_df['Адрес места'].apply(extract_city_and_floor)
cleaned_df[['Город', 'Этаж']] = new_columns

cleaned_df.head()

updated_file_path_with_floor = 'zoon_data_with_city_and_floor.csv'
cleaned_df.to_csv(updated_file_path_with_floor, index=False, encoding='UTF-8-SIG')
```

Удалить ссылку, отделить этаж, город и адрес, убрать мин из метро и записать их в название столбца, сформировать новый признак на основе оценки и количестве пользователей, делаем сплит по подкатегориям,

```python
cleaned_df = cleaned_df.drop(['Ссылка'], axis=1)
```

```python
unique_time_values = cleaned_df['Время до метро'].unique()
print(unique_time_values)
```

```python
def clean_time(value):
    if isinstance(value, str):
        if 'мин.' in value:
            return float(value.replace('мин.', '').strip())
        elif 'км' in value:
            return np.nan
        elif value == 'Нет данных':
            return np.nan
    return np.nan

cleaned_df['Время до метро'] = cleaned_df['Время до метро'].apply(clean_time)

cleaned_df.rename(columns={'Время до метро': 'Время до метро в минутах'}, inplace=True)

```

```python
cleaned_df.head()
```

```python
cleaned_df['Отзывы'] = pd.to_numeric(cleaned_df['Отзывы'], errors='coerce')


cleaned_df['Релевантная оценка'] = cleaned_df['Рейтинг'] * cleaned_df['Отзывы']

updated_file_path_final = 'zoon_data_final.csv'
cleaned_df.to_csv(updated_file_path_final, index=False, encoding='UTF-8-SIG')

cleaned_df.head()
```
